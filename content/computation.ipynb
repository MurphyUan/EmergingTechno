Growth Rates & Big O Notation
====

For the purposes of understanding **Growth Rates and Big-O notation**, we were tasked with plotting 5 five pairs of **functions f(x) and g(x)**, such that **f(x) is O(g(x))** (read as "big-O of g"). This explores the *relative running time of programs* based on their **complexity and efficiency**, these can be easily represented as *polynomial functions* these are denoted, with their *Big-O counterpart* as follows: $$f(x) = 3x^2 + 5x - 7$$ $$g(x) = x^2$$

You could be wondering how we came to this conclusion, if we take our two functions we can show that for function *$f(x)$* is *$O(g(x))$* where there are *positive integers* *$c$* and *$x_0$* such that for every $x \ge x_0$: $f(x) \le cg(x)$. Now how do we do we show that this is true, first we choose values for  $c$ and $x_0$, then we must show that $0 \le f(x) \le\le cg(x)$ for every $x \ge x_0$.

For the above example, we choose $c = 4$ and $x_0 = 100$. For any $x \ge x_0$, $x^2 \ge 100x$ is also true as we multiply both sides by x, we can also see that $100x \ge 5x - 7$ making $x^2 \ge 5x - 7$. Since x is positive we can say that $0 \le 3x^2 + 5x - 7$. Therefore for any $x \ge k$, $0 \le 3x^2 + 5x - 7 \le 3x^2 + x^2 = 4x^2 = cx^2$. Therefore $0 \le f(x) \le\le cg(x)$ for every $x \ge x_0$.

```python
import matplotlib.pyplot as plt
import numpy as np
import math

%matplotlib inline

plt.rcParams["figure.figsize"] = [12,8]
x = np.linspace(0, 100, 50)

f0 = 4*x + 7
g0 = x

f1 = 3*x**2 + 5*x - 7
g1 = x**2

f2 = 2*x**3 - 2*x**2 + 3*x + 10
g2 = x**3

f3 = 3*(np.log(x))
g3 = np.log(x)

f4 = 3*x*(np.log(x))
g4 = x*(np.log(x))

plt.subplot(2,3,1)
plt.plot(x, f0, 'yx-', label = "")
plt.plot(x, g0, 'g')

plt.subplot(2,3,2)
plt.plot(x, f1, 'yx-')
plt.plot(x, g1, 'g')

plt.subplot(2,3,3)
plt.plot(x, f2, 'yx-')
plt.plot(x, g2, 'g')

plt.subplot(2,3,4)
plt.plot(x, f3, 'yx-')
plt.plot(x, g3, 'g')

plt.subplot(2,3,5)
plt.plot(x, f4, 'yx-')
plt.plot(x, g4, 'g')
plt.show()

```

Turing Machine
======

```python
#State table
table = {
    ('X','_'):['_','R','T'],
    ('X','0'):['0','R','X'],
    ('X','1'):['1','R','F'],
}

# Tape input
tape = list('0000010000')
# Position on tape
pos = 0
# Initial state is first in table
state = 'X'

# Keep going while we are not in a halting state.
while state not in ['T', 'F']:
    # Print the current status.
    print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
    # Get the row of the table.
    row = table[(state, tape[pos])]
    # Overwrite the symbol.
    tape[pos] = row[0]
    # Move left or right.
    if row[1] == 'R':
        # Put blanks on tape as necessary.
        if pos == len(tape) - 1:
            tape = tape + ['_']
        # Increase position.
        pos = pos + 1
    else:
        # Put blanks on tape as necessary.
        if pos == 0:
            tape = ['_'] + tape
            # The position on the tape has to move with it.
            pos = pos + 1
        # Increase position.
        pos = pos - 1
    # Update the state.
    state = row[2]

# Print the current status.
print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
```
